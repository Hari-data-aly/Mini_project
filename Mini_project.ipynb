```python
# Aggregate sales by month and year
cursor.execute('''
SELECT
    strftime('%Y', order_date) AS year,
    strftime('%m', order_date) AS month,
    SUM(sales_amount) AS total_sales
FROM
    Montly_sales
GROUP BY
    year, month
ORDER BY
    year, month
''')

# Fetch all rows
rows = cursor.fetchall()

# Display the rows
for row in rows:
    print(row)
```

```python
# Calculate year-over-year growth
cursor.execute('''
WITH monthly_sales AS (
    SELECT
        strftime('%Y', order_date) AS year,
        strftime('%m', order_date) AS month,
        SUM(sales_amount) AS total_sales
    FROM
        Montly_sales
    GROUP BY
        year, month
)
SELECT
    a.year AS current_year,
    a.month,
    a.total_sales AS current_year_sales,
    b.total_sales AS previous_year_sales,
    ((a.total_sales - b.total_sales) / b.total_sales) * 100 AS yoy_growth
FROM
    monthly_sales a
LEFT JOIN
    monthly_sales b
ON
    a.month = b.month AND a.year = b.year + 1
ORDER BY
    a.year, a.month
''')

# Fetch all rows
rows = cursor.fetchall()

# Display the rows
for row in rows:
    print(row)
```

```python
# Calculate revenue and profit margin
cursor.execute('''
SELECT
    order_id,
    Category,
    sub_category,
    Quantity,
    cost_price,
    (Quantity * cost_price) AS revenue,
    ((Quantity * cost_price) - (Quantity * cost_price)) AS profit,
    (((Quantity * cost_price) - (Quantity * cost_price)) / (Quantity * cost_price)) * 100 AS profit_margin
FROM
    sales
''')

# Fetch all rows
rows = cursor.fetchall()

# Display the rows
for row in rows:
    print(row)
```

```python
# Categorize products by profit margin
cursor.execute('''
SELECT
    order_id,
    Category,
    sub_category,
    Quantity,
    cost_price,
    (Quantity * cost_price) AS revenue,
    ((Quantity * cost_price) - (Quantity * cost_price)) AS profit,
    (((Quantity * cost_price) - (Quantity * cost_price)) / (Quantity * cost_price)) * 100 AS profit_margin,
    CASE
        WHEN (((Quantity * cost_price) - (Quantity * cost_price)) / (Quantity * cost_price)) * 100 >= 50 THEN 'High'
        WHEN (((Quantity * cost_price) - (Quantity * cost_price)) / (Quantity * cost_price)) * 100 >= 20 THEN 'Medium'
        ELSE 'Low'
    END AS profit_category
FROM
    sales
''')

# Fetch all rows
rows = cursor.fetchall()

# Display the rows
for row in rows:
    print(row)
```

```python
# Rank products by revenue
cursor.execute('''
SELECT
    order_id,
    Category,
    sub_category,
    Quantity,
    cost_price,
    (Quantity * cost_price) AS revenue,
    ROW_NUMBER() OVER (ORDER BY (Quantity * cost_price) DESC) AS revenue_rank
FROM
    sales
''')

# Fetch all rows
rows = cursor.fetchall()

# Display the rows
for row in rows:
    print(row)
```

```python
# Filter products with high revenue
cursor.execute('''
SELECT
    order_id,
    Category,
    sub_category,
    Quantity,
    cost_price,
    (Quantity * cost_price) AS revenue
FROM
    sales
GROUP BY
    order_id, Category, sub_category, Quantity, cost_price
HAVING
    (Quantity * cost_price) > 2000
''')

# Fetch all rows
rows = cursor.fetchall()

# Display the rows
for row in rows:
    print(row)
```

```python
# Find the total profit per category
cursor.execute('''
SELECT
    c.category_name,
    SUM((r.sale_price - r.cost_price) * r.quantity_sold) AS total_profit
FROM
    categories c
JOIN
    goods g ON c.category_id = g.category_id
JOIN
    records r ON g.good_id = r.good_id
GROUP BY
    c.category_name
''')

# Fetch all rows
rows = cursor.fetchall()

# Display the rows
for row in rows:
    print(row)
```

```python
# Determine the average discount percentage given per region
cursor.execute('''
SELECT
    r.region_name,
    AVG((l.original_price - l.sale_price) / l.original_price * 100) AS average_discount_percentage
FROM
    regions r
JOIN
    inventory i ON r.region_id = i.region_id
JOIN
    logs l ON i.inventory_id = l.inventory_id
GROUP BY
    r.region_name
''')

# Fetch all rows
rows = cursor.fetchall()

# Display the rows
for row in rows:
    print(row)
```
